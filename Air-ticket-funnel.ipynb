## 2. Общий анализ воронки

```python
stage_order = [
    'Поиск',
    'Выдача',
    'Оффер',
    'Чекаут',
    'Успешная покупка'
]

stage_values = [
    df.loc[df['screen_nm'] == 'Поиск', 'session_id'].nunique(),
    df.loc[df['screen_nm'] == 'Выдача', 'session_id'].nunique(),
    df.loc[df['screen_nm'] == 'Оффер', 'session_id'].nunique(),
    df.loc[df['screen_nm'] == 'Чекаут', 'session_id'].nunique(),
    df.loc[df['session_purchase_status'] == 'Успешные', 'session_id'].nunique()
]

funnel_df = pd.DataFrame({
    'stage': stage_order,
    'sessions': stage_values
})

fig = go.Figure(go.Funnel(
    y=funnel_df['stage'],
    x=funnel_df['sessions'],
    textinfo='value+percent initial+percent previous'
))

fig.update_layout(
    title='Воронка покупки авиабилетов',
    xaxis_title='Количество уникальных сессий',
    yaxis_title='Этап'
)

fig.show()
```

На этапе "Поиск" наблюдается 37 718 уникальных сессий, что соответствует 100% от общего количества. На следующем этапе "Выдача" количество сессий сокращается до 34 286, что составляет 91% от начального этапа. Далее, на этапе "Оффер", количество сессий уменьшается до 17 703, что составляет 46.9% от начального этапа и 51.6% от предыдущего этапа. На этапе "Чекаут" количество сессий сокращается до 6466, что составляет 17.1% от начального этапа и 36.5% от предыдущего этапа. Наконец, на этапе "Покупка" количество сессий составляет 2362, что соответствует 6.3% от начального этапа и 36.5% от предыдущего этапа. 

### Таблица потерь по шагам воронки

```python
funnel_table = funnel_df.copy()

funnel_table['drop_from_prev'] = funnel_table['sessions'].shift(1) - funnel_table['sessions']
funnel_table.loc[0, 'drop_from_prev'] = 0

funnel_table['cr_from_prev'] = funnel_table['sessions'] / funnel_table['sessions'].shift(1)
funnel_table.loc[0, 'cr_from_prev'] = 1

funnel_table['cr_from_start'] = funnel_table['sessions'] / funnel_table['sessions'].iloc[0]

funnel_table['drop_pct_from_prev'] = 1 - funnel_table['cr_from_prev']
funnel_table.loc[0, 'drop_pct_from_prev'] = 0

percent_cols = ['cr_from_prev', 'cr_from_start', 'drop_pct_from_prev']
count_cols = ['sessions', 'drop_from_prev']

funnel_table[percent_cols] = (funnel_table[percent_cols] * 100).round(2)
funnel_table[count_cols] = funnel_table[count_cols].astype(int)

funnel_table
```

По таблице потерь видно, что основная абсолютная потеря пользователей происходит на переходе `Выдача → Оффер`: на этом шаге воронка теряет 16.6 тыс. сессий, или 48.4% пользователей предыдущего этапа. Это делает экран выдачи одной из ключевых точек для продуктовой оптимизации.

При этом наиболее низкая относительная конверсия наблюдается в нижней части воронки: переходы `Оффер → Чекаут` и `Чекаут → Успешная покупка` имеют конверсию около 36.5%, то есть на каждом из этих шагов теряется почти две трети пользователей. В результате итоговая конверсия от поиска до успешной покупки составляет 6.26%.

## 3. Углубление в проблемный шаг Выдача -> Оффер

```python
screen_times = (
    df[df['screen_nm'].isin(['Выдача', 'Оффер'])]
    .sort_values(['session_id', 'screen_dttm'])
    .groupby(['session_id', 'screen_nm'], as_index=False)['screen_dttm']
    .first()
    .pivot(index='session_id', columns='screen_nm', values='screen_dttm')
    .reset_index()
)
screen_times = screen_times[screen_times['Выдача'].notna()].copy()
screen_times['reached_offer'] = (
        screen_times['Оффер'].notna() &
        (screen_times['Оффер'] >= screen_times['Выдача'])
)

last_screen_time = (
    df.groupby('session_id', as_index=False)['screen_dttm']
    .max()
    .rename(columns={'screen_dttm': 'last_screen_dttm'})
)

screen_times = screen_times.merge(last_screen_time, on='session_id', how='left')

screen_times['time_on_results_sec'] = np.where(
    screen_times['reached_offer'],
    (screen_times['Оффер'] - screen_times['Выдача']).dt.total_seconds(),
    (screen_times['last_screen_dttm'] - screen_times['Выдача']).dt.total_seconds()
)

screen_times = screen_times[screen_times['time_on_results_sec'] > 0].copy()

bins = [-np.inf, 10, 30, 60, 120, 300, 600, 1200, 3600, np.inf]
labels = ['<=10с', '10–30с', '30–60с', '1–2м', '2–5м', '5–10м', '10–20м', '20–60м', '60м+']

screen_times['time_bucket'] = pd.cut(
    screen_times['time_on_results_sec'],
    bins=bins,
    labels=labels,
    right=True
)
bucket_stats = (
    screen_times.groupby('time_bucket', observed=False).agg(sessions_cnt=('session_id', 'nunique'), reached_offer_share=('reached_offer', 'mean')).reset_index())

bucket_stats
```

```python
fig = px.bar(
    bucket_stats,
    x='time_bucket',
    y='reached_offer_share',
    text='sessions_cnt',
    title='Переход Выдача -> Оффер в зависимости от времени на Выдаче',
    labels={
        'time_bucket': 'Время на Выдаче',
        'reached_offer_share': 'Доля сессий, дошедших до Оффера'
    }
)

fig.update_yaxes(range=[0, 1.08], tickformat='.0%')
fig.update_traces(
    texttemplate='n=%{text}',
    textposition='outside'
)

fig.update_layout(
    xaxis_title='Время на Выдаче',
    yaxis_title='Доля сессий, дошедших до Оффера'
)

fig.show()
```

По графику видно, что пользователи, которые быстро переходят с выдачи к офферу, чаще доходят до следующего шага. А если пользователь проводит на выдаче много времени, вероятность перехода в оффер снижается.

Это может говорить о том, что часть пользователей быстро находит подходящий вариант и сразу открывает его. А те, кто долго остаётся на выдаче, скорее сталкиваются со сложностью выбора или не находят достаточно подходящее предложение.

Значит, экран выдачи можно считать одной из важных точек для улучшения продукта.

### Время перехода от чекаута к успешной покупке
Посмотрим, сколько времени в среднем пользователи тратят на переход от этапа `Чекаут` к успешной покупке.

```python
checkout_purchase_times = (
    df[df['screen_nm'] == 'Чекаут']
    .sort_values(['session_id', 'screen_dttm'])
    .groupby('session_id', as_index=False)['screen_dttm']
    .first()
    .rename(columns={'screen_dttm': 'checkout_time'})
)

purchase_times = (
    df[df['session_purchase_status'] == 'Успешные']
    .sort_values(['session_id', 'screen_dttm'])
    .groupby('session_id', as_index=False)['screen_dttm']
    .max()
    .rename(columns={'screen_dttm': 'purchase_time'})
)

checkout_purchase_times = checkout_purchase_times.merge(
    purchase_times,
    on='session_id',
    how='inner'
)

checkout_purchase_times = checkout_purchase_times[
    checkout_purchase_times['purchase_time'] >= checkout_purchase_times['checkout_time']
].copy()

checkout_purchase_times['checkout_to_purchase_sec'] = (
    checkout_purchase_times['purchase_time'] - checkout_purchase_times['checkout_time']
).dt.total_seconds()

checkout_purchase_summary = pd.DataFrame({
    'metric': ['mean_sec', 'median_sec', 'mean_min', 'median_min'],
    'value': [
        checkout_purchase_times['checkout_to_purchase_sec'].mean(),
        checkout_purchase_times['checkout_to_purchase_sec'].median(),
        checkout_purchase_times['checkout_to_purchase_sec'].mean() / 60,
        checkout_purchase_times['checkout_to_purchase_sec'].median() / 60,
    ]
})

checkout_purchase_summary['value'] = checkout_purchase_summary['value'].round(2)
checkout_purchase_summary
```

По таблице видно, что пользователи, которые дошли до чекаута и завершили покупку, обычно делают это довольно быстро: медианное время от чекаута до покупки составляет около 6.7 минуты.

При этом среднее время заметно выше — около 35 минут. Это значит, что в данных есть часть длинных сессий, которые увеличивают среднее значение. Поэтому медиана здесь лучше показывает типичное поведение пользователя.

В целом можно сказать, что успешная покупка после перехода в чекаут чаще всего происходит в течение нескольких минут, но у части пользователей процесс заметно затягивается.

```python
sessions_by_month = (
    df.groupby('month')['session_id']
    .nunique()
    .reset_index()
    .rename(columns={'session_id': 'sessions_cnt'})
)
fig = px.bar(
    sessions_by_month,
    x='month',
    y='sessions_cnt',
    title='Количество сессий по месяцам',
    labels={
        'month': 'Месяц',
        'sessions_cnt': 'Количество сессий'
    }
)
fig.update_traces(textposition='outside')
fig.show()
```

Количество пользователей и сессий по месяцам в целом стабильное, посмотрим количество покупок и конверсию в покупку по месяцам, чтобы выявить возможные аномалии.

```python
purchases_by_month = (
    df.loc[df['session_purchase_status'] == 'Успешные']
    .groupby('month')['session_id']
    .nunique()
    .reset_index()
    .rename(columns={'session_id': 'purchases_cnt'})
)

conversion_by_month = sessions_by_month.merge(
    purchases_by_month,
    on='month',
    how='left'
)

conversion_by_month['purchases_cnt'] = conversion_by_month['purchases_cnt'].fillna(0)
conversion_by_month['conversion_to_purchase'] = (
    conversion_by_month['purchases_cnt'] / conversion_by_month['sessions_cnt']
)

conversion_by_month
```

```python
monthly_metrics = sessions_by_month.merge(
    purchases_by_month,
    on='month',
    how='left'
)

monthly_metrics['purchases_cnt'] = monthly_metrics['purchases_cnt'].fillna(0)
monthly_metrics['conversion_to_purchase'] = (
    monthly_metrics['purchases_cnt'] / monthly_metrics['sessions_cnt']
)
monthly_metrics['conversion_pct'] = (
    monthly_metrics['conversion_to_purchase'] * 100
).round(2)

fig = px.bar(
    monthly_metrics,
    x='month',
    y='purchases_cnt',
    title='Количество покупок по месяцам',
    labels={
        'month': 'Месяц',
        'purchases_cnt': 'Количество покупок'
    },
    text='purchases_cnt'
)

fig.update_traces(textposition='outside')
fig.show()
```

```python
fig = px.line(
    monthly_metrics,
    x='month',
    y='conversion_to_purchase',
    markers=True,
    title='Конверсия в покупку по месяцам',
    labels={
        'month': 'Месяц',
        'conversion_to_purchase': 'Конверсия в покупку'
    }
)

fig.update_yaxes(tickformat='.1%')
fig.show()
```

Анализ динамики покупок и конверсии по месяцам показал, что в большинстве месяцев метрики оставались относительно стабильными. Однако в июле наблюдается резкое снижение как количества покупок, так и конверсии в покупку: число покупок сократилось до 82, а конверсия — до 2.0%. Поскольку одновременно снизились и абсолютный объем покупок, и эффективность перевода сессий в покупку, июль можно считать аномальным месяцем.

```python
def build_funnel(data, label):
    result = pd.DataFrame({
        'stage': stage_order,
        'sessions': [
            data.loc[data['screen_nm'] == 'Поиск', 'session_id'].nunique(),
            data.loc[data['screen_nm'] == 'Выдача', 'session_id'].nunique(),
            data.loc[data['screen_nm'] == 'Оффер', 'session_id'].nunique(),
            data.loc[data['screen_nm'] == 'Чекаут', 'session_id'].nunique(),
            data.loc[data['session_purchase_status'] == 'Успешные', 'session_id'].nunique()
        ]
    })
    result['group'] = label
    result['cr_from_start'] = result['sessions'] / result['sessions'].iloc[0]
    result['cr_from_prev'] = result['sessions'] / result['sessions'].shift(1)
    result.loc[0, 'cr_from_prev'] = 1
    return result

july = build_funnel(df[df['month'] == pd.Timestamp('2025-07-01')], 'Июль')
august = build_funnel(df[df['month'] == pd.Timestamp('2025-08-01')], 'Август')
other = build_funnel(
    df[~df['month'].isin([pd.Timestamp('2025-07-01'), pd.Timestamp('2025-08-01')])],
    'Остальные месяцы'
)

funnel_compare = pd.concat([july, august, other], ignore_index=True)

fig = px.bar(
    funnel_compare,
    x='stage',
    y='cr_from_start',
    color='group',
    barmode='group',
    title='Сравнение конверсии по шагам: июль, август и остальные месяцы',
    labels={
        'stage': 'Этап воронки',
        'cr_from_start': 'Конверсия от первого шага',
        'group': 'Период'
    },
    text='cr_from_start'
)

fig.update_yaxes(tickformat='.1%')
fig.update_traces(texttemplate='%{text:.1%}', textposition='outside')
fig.show()
```

График показывает, что проблема июля сосредоточена в нижней части воронки. Верхний этап (поиск -> выдача) просел незначительно, однако начиная с оффера июль заметно отстаёт от августа и остальных месяцев. Особенно сильная разница видна на этапах чекаута и успешной покупки: итоговая конверсия июля составляет 2.1% против 6.9% в августе и 6.7% в остальные месяцы. Это указывает на локальную проблему июля в процессе оформления или покупки, а не на снижение объёма или качества входящего трафика.

## 5. Сегментация конверсии по каналам привлечения

### Конверсия в покупку по session_channel

```python
channel_sessions = (
    df.groupby('session_channel')['session_id']
    .nunique()
    .reset_index()
    .rename(columns={'session_id': 'sessions_cnt'})
)

channel_purchases = (
    df.loc[df['session_purchase_status'] == 'Успешные']
    .groupby('session_channel')['session_id']
    .nunique()
    .reset_index()
    .rename(columns={'session_id': 'purchases_cnt'})
)

channel_conversion = channel_sessions.merge(
    channel_purchases,
    on='session_channel',
    how='left'
)

channel_conversion['purchases_cnt'] = channel_conversion['purchases_cnt'].fillna(0)
channel_conversion['conversion_to_purchase'] = (
    channel_conversion['purchases_cnt'] / channel_conversion['sessions_cnt']
)
channel_conversion = channel_conversion.sort_values('conversion_to_purchase', ascending=False)

channel_conversion
```

```python
fig = px.bar(
    channel_conversion,
    x='session_channel',
    y='conversion_to_purchase',
    title='Конверсия в покупку по session_channel',
    labels={
        'session_channel': 'Канал',
        'conversion_to_purchase': 'Конверсия в покупку'
    },
    text='conversion_to_purchase'
)

fig.update_yaxes(tickformat='.1%', range=[0, 0.08])

fig.update_traces(
    texttemplate='%{text:.1%}',
    textposition='outside',
    textfont_size=14,
    width=0.7
)

fig.update_layout(
    height=650,
    xaxis_tickangle=-35,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
    margin=dict(t=120, b=120, l=80, r=40)
)

fig.show()
```

Конверсия в покупку по session_channel различается умеренно - от 5.4% до 7.2%. Наиболее высокий уровень конверсии наблюдается у banner, Google и email, тогда как наиболее низкий - у Яндекса и Сторис. При этом каналы существенно различаются по объёму трафика. На текущем этапе можно предположить, что отдельные каналы различаются по качеству привлечения, однако для уверенного вывода необходимо дополнительно проверить статистическую значимость различий и отдельно проанализировать поведение каналов в июле.

### Динамика средней цены покупки по месяцам

Анализ показал, что начиная с июля средняя и медианная цена успешной покупки заметно выросли относительно первой половины года. В июле средняя цена составила 16.8 тыс., тогда как в январе–июне она находилась в диапазоне примерно 13.6–15.0 тыс. При этом в августе и сентябре уровень цен сохранился и даже оказался выше июльского.

Это означает, что июльская просадка конверсии не может быть объяснена только ростом цены. Если бы основной причиной было исключительно удорожание покупки, то снижение конверсии должно было бы сохраняться и в августе–сентябре, когда средняя цена оставалась высокой. Однако после июля конверсия восстановилась.

Следовательно, рост среднего чека мог быть сопутствующим фактором, но не выглядит основной причиной июльской аномалии. Для дальнейшего анализа необходимо проверить не только уровень цены, но и структуру покупок: число билетов, направления, класс обслуживания, продавцов и другие характеристики офферов.

### Динамика количества билетов в покупке по месяцам

Анализ показал, что начиная с мая–июля в успешных покупках растёт среднее количество билетов. Если в январе–апреле среднее значение находилось в диапазоне 1.46–1.52, то в июле оно достигло 1.68, а в августе–сентябре — 1.75–1.78. При этом медианное значение на всём периоде остаётся равным 1, что указывает на сохранение доминирующего сценария покупки одного билета; рост среднего значения, вероятно, связан с увеличением доли заказов с несколькими билетами.

Это означает, что рост среднего чека с июля частично может объясняться не только изменением уровня цен, но и увеличением числа билетов в одной покупке. Однако сам по себе этот фактор не объясняет июльскую просадку конверсии: в августе и сентябре среднее количество билетов остаётся ещё выше, тогда как конверсия в покупку восстанавливается.

Следовательно, изменение размера заказа могло сопровождать июльскую аномалию, но не выглядит её основной причиной.

## 7. Сводка ключевых метрик

```python
monthly_summary = (
    sessions_by_month
    .merge(users_by_month, on='month', how='left')
    .merge(purchases_by_month, on='month', how='left')
    .merge(
        purchase_price_by_month[['month', 'avg_purchase_price', 'median_purchase_price']],
        on='month',
        how='left'
    )
    .merge(
        tickets_by_month[['month', 'avg_tickets_num', 'median_tickets_num']],
        on='month',
        how='left'
    )
)

monthly_summary['purchases_cnt'] = monthly_summary['purchases_cnt'].fillna(0)
monthly_summary['conversion_to_purchase'] = (
    monthly_summary['purchases_cnt'] / monthly_summary['sessions_cnt']
)

monthly_summary['conversion_pct'] = (
    monthly_summary['conversion_to_purchase'] * 100
).round(2)

monthly_summary = monthly_summary[
    [
        'month',
        'users_cnt',
        'sessions_cnt',
        'purchases_cnt',
        'conversion_pct',
        'avg_purchase_price',
        'median_purchase_price',
        'avg_tickets_num',
        'median_tickets_num'
    ]
]

monthly_summary
```

## 9. Продуктовые гипотезы

На основе проведённого анализа я сформулировал 3 продуктовые гипотезы, направленные на улучшение ключевых этапов клиентской воронки.

### Гипотеза 1. Улучшение экрана выдачи может повысить переход в оффер
Переход `Выдача → Оффер` даёт крупнейшую абсолютную потерю пользователей в воронке. Кроме того, сессии с длительным временем на экране выдачи заметно реже доходят до оффера, чем сессии с более быстрым переходом.
Если на экране выдачи визуально выделять наиболее релевантные и выгодные предложения, то пользователям будет проще ориентироваться в результатах поиска, и конверсия `Выдача → Оффер` может вырасти.

**Продуктовое решение:**
Добавить персонализированную подборку, рекомендательный блок или визуально выделить наиболее релевантные и выгодные варианты в верхней части выдачи.

**Ожидаемый эффект:**
- рост CR `Выдача → Оффер`;
- снижение доли быстрых выходов с экрана выдачи;
- рост итоговой конверсии в покупку.

### Гипотеза 2. Возвратные коммуникации для пользователей, не завершивших покупку, повысят конверсию
Существенная часть пользователей доходит до этапов выдачи и оффера, но не завершает покупку в рамках текущей сессии.
Если запускать персонализированные пуши, письма или ретаргетинговые коммуникации для пользователей, которые дошли до выдачи или оффера, но не завершили покупку, то часть из них вернётся в воронку и завершит заказ.

**Продуктовое решение:**
Запустить персонализированные напоминания и ретаргетинговые коммуникации для пользователей, которые были на этапе выдачи или оффера, но не завершили покупку.

**Ожидаемый эффект:**
- рост доли возвратов в воронку;
- увеличение числа покупок без пропорционального роста затрат на привлечение.

### Гипотеза 3. Улучшение условий оффера в сезонный период может повысить конверсию в покупку
В июле 2025 года основная просадка наблюдается не в верхней части воронки, а начиная с этапа оффера и дальше — на шагах оформления и покупки.
Если в период сезонного спроса предлагать пользователям более конкурентные условия — например, специальные предложения, скидки на чувствительные сегменты или более заметную коммуникацию выгодных офферов, — то переход из оффера в чекаут и итоговая конверсия в покупку могут вырасти.

**Продуктовое решение:**
В сезонные периоды выделять наиболее выгодные предложения, тестировать специальные условия для чувствительных сегментов и усиливать коммуникацию ценности оффера на этапе выбора.

**Ожидаемый эффект:**
- рост CR `Оффер → Чекаут`;
- рост итоговой конверсии в покупку;
- снижение сезонной просадки конверсии.

### Наиболее перспективная гипотеза

Наиболее перспективной выглядит **гипотеза об улучшении экрана выдачи**.

Именно эта гипотеза выглядит наиболее перспективной, потому что она напрямую опирается на главный вывод исследования: крупнейшая абсолютная потеря пользователей происходит на переходе `Выдача → Оффер`. Кроме того, дополнительный анализ времени на выдаче показал, что длительное взаимодействие с этим этапом связано с более низкой вероятностью перехода дальше по воронке. Улучшение сценария выбора на выдаче потенциально способно повлиять сразу на большой объём трафика и повысить эффективность всей воронки, а не только её финального этапа.

**Как я бы проверял эту гипотезу:**
Я бы запустил A/B-тест, в котором пользователи случайным образом делятся на контрольную и тестовую группы. В контрольной группе экран выдачи остаётся без изменений, а в тестовой добавляется улучшенный сценарий — например, визуальное выделение лучших предложений, рекомендательный блок или более заметная подача выгодных вариантов.

Основной метрикой эксперимента я бы выбрал конверсию `Выдача → Оффер`, так как именно этот шаг напрямую отражает эффект изменения интерфейса. В качестве дополнительных метрик я бы отслеживал итоговую конверсию в покупку, долю быстрых выходов с экрана выдачи и возможные негативные эффекты, например увеличение времени выбора без роста переходов дальше по воронке.

Если в тестовой группе переход `Выдача → Оффер` и итоговая конверсия окажутся выше без ухудшения пользовательского поведения на следующих шагах, гипотезу можно будет считать подтверждённой.
